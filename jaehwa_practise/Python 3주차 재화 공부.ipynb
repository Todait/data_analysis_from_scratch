### 파일 불러오기

```python
import pandas as pd
```

```python
test = pd.read_csv("test.csv", index_col="PassengerId")
print(test.shape)
test.head()
```

```python
train = pd.read_csv("train.csv", index_col="PassengerId")
print(train.shape)
train.head()
```

## Encode sex

```python
test.loc[test["Sex"] == "male", "Sex_encode"] = 0
test.loc[test["Sex"] == "female", "Sex_encode"] = 1

print(test.shape)
test.head()

test[["Sex", "Sex_encode"]].head()
```

```python
train.loc[train["Sex"] == "male", "Sex_encode"] = 0
train.loc[train["Sex"] == "female", "Sex_encode"] = 1

print(train.shape)
train.head()

train[["Sex", "Sex_encode"]].head()
```

## 열에  null 체킹

```python
train[pd.isnull(train["Age"])].head()
```

```python
mean_fare = train["Age"].mean()
print(mean_fare)
```

## 연습문제 1 - 나이의 빈곳은 평균값으로 채운 상태에서의 나이의 평균값들을 계산해보자

```python
train.loc[pd.notnull(train["Age"]), "Age_fillin"] = train["Age"]
train.loc[pd.isnull(train["Age"]), "Age_fillin"] = train["Age"].mean()

train[["Age", "Age_fillin"]].head()
```

```python
print(train["Age_fillin"].mean())
```

## one hot incoding

```python
test_embarked = pd.get_dummies(test["Embarked"], prefix="Embarked")

test = pd.concat([test, test_embarked], axis=1)

test[["Embarked_Q","Embarked_S","Embarked_C", "Embarked"]].head()
```

```python
train_embarked = pd.get_dummies(train["Embarked"], prefix="Embarked")

train = pd.concat([train, train_embarked], axis=1)

train[["Embarked_Q","Embarked_S","Embarked_C", "Embarked"]].head()
```

## 연습문제 2 Mr, Mrs, Miss, Ms, Master 가 들어간 이름에 나이의 평균값을 Age 가  없는 빈칸에 채워넣어라
### 단, 반복되는 코드는 함수를 사용해서 처리하라

```python
def get_age_by_name_title(title):
    true_coulmn = train[train["Name"].str.contains(title)]
    mean = true_coulmn["Age"].mean()
    return mean

def set_age_fillin_by_name_title(title):
    train.loc[pd.isnull(train["Age"]) & train["Name"].str.contains(title), "Age_fillin"] = get_age_by_name_title(title)
    return
    
set_age_fillin_by_name_title("Mr\.")
set_age_fillin_by_name_title("Mrs\.")
set_age_fillin_by_name_title("Miss\.")
set_age_fillin_by_name_title("Ms\.")
set_age_fillin_by_name_title("Master\.")

print("일반 평균 : {0}".format(train["Age"].mean()))
print("이름 채운 후 평균 : {0}".format(train["Age_fillin"].mean()))
```

```python
train["Age_fillin"]
```

```python
train[pd.isnull(train["Age_fillin"])].head()
```

```python
test.loc[pd.notnull(test["Age"]), "Age_fillin"] = test["Age"]

def get_age_by_name_title_in_test(title):
    true_coulmn = test[test["Name"].str.contains(title)]
    mean = true_coulmn["Age"].mean()
    return mean

def set_age_fillin_by_name_title_test(title):
    test.loc[pd.isnull(test["Age"]) & test["Name"].str.contains(title), "Age_fillin"] = get_age_by_name_title_in_test(title)
    return

set_age_fillin_by_name_title_test("Mr\.")
set_age_fillin_by_name_title_test("Mrs\.")
set_age_fillin_by_name_title_test("Miss\.")
set_age_fillin_by_name_title_test("Ms\.")
set_age_fillin_by_name_title_test("Master\.")

print("일반 평균 : {0}".format(test["Age"].mean()))
print("이름 채운 후 평균 : {0}".format(test["Age_fillin"].mean()))
```

```python
test[pd.isnull(test["Age_fillin"])] = test["Age"].mean() 
## 여기에서 test[pd.isnull(test["Age_fillin"])] 을 사용시 Age_fillin이 채워지지 않은 하나가 나옴 -> 아직 이유는 못찾음
```

## Train


```python
train["Fare_fillin"] = train["Fare"]
test["Fare_fillin"] = test["Fare"]

test.loc[pd.isnull(test["Fare"]), "Fare_fillin"] = mean_fare

missing_fare = test[pd.isnull(test["Fare"])]

print(missing_fare.shape)

missing_fare.index

test.loc[missing_fare.index, ["Fare", "Fare_fillin"]]
```

```python
feature_name = ["Pclass", "Sex_encode","Fare_fillin",
                "Embarked_C", "Embarked_Q", "Embarked_S","Age_fillin"]
X_train = train[feature_name]
print(X_train.shape)
X_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth = 3, random_state=37)
```

```python
label_name = "Survived"

y_train = train[label_name]

print(y_train.shape)
y_train.head()
```

```python
model.fit(X_train, y_train)
```

```python
X_test = test[feature_name]
print(X_test)
```

```python
prediction = model.predict(X_test)
submission = pd.read_csv("gender_submission.csv", index_col="PassengerId")
submission["Survived"] = prediction

print(submission.shape)
submission
```

```python
submission.to_csv("baseline-script111.csv")
```
